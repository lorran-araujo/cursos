```python
from sklearn.decomposition import KernelPCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import necessário para plotagem 3D

# Carregar dados de exemplo (Iris)
X, y = load_iris(return_X_y=True)

# Aplicar Kernel PCA com o kernel RBF para reduzir para 3 componentes
kpca = KernelPCA(n_components=3, kernel='rbf', gamma=15)
X_kpca = kpca.fit_transform(X)

# Criação de figura 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plotando os dados transformados em 3D
scatter = ax.scatter(X_kpca[:, 0], X_kpca[:, 1], X_kpca[:, 2], c=y, cmap=plt.cm.viridis)

# Rótulos dos eixos
ax.set_xlabel('Componente Principal 1')
ax.set_ylabel('Componente Principal 2')
ax.set_zlabel('Componente Principal 3')
ax.set_title('Kernel PCA com kernel RBF em 3D')

# Exibir a legenda
legend1 = ax.legend(*scatter.legend_elements(), title="Classes")
ax.add_artist(legend1)

# Mostrar o gráfico
plt.show()
```
